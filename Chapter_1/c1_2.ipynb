### Python 中的字串操作

#### 1. 字串創建

您可以使用單引號、雙引號或三重引號（用於多行字串）來創建字串。

```python
single_quote_string = 'Hello, World!'
double_quote_string = "Hello, World!"
triple_quote_string = '''這是一個 
多行字串。'''
```

#### 2. 字串連接

您可以使用 `+` 運算符來連接（加入）字串。

```python
str1 = "Hello"
str2 = "World"
result = str1 + " " + str2  # "Hello World"
print(result)
```

#### 3. 字串重複

您可以使用 `*` 運算符來重複字串。

```python
str3 = "Hello "
result = str3 * 3  # "Hello Hello Hello "
print(result)
```

#### 4. 字串長度

您可以使用 `len()` 函數來獲取字串的長度。

```python
length = len("Hello")  # 5
print("字串的長度:", length)
```

#### 5. 字串方法

Python 提供了幾個內建方法來進行字串操作。以下是一些常用的方法：

- **`.lower()`**：將字串轉換為小寫。

  ```python
  lower_string = "Hello".lower()  # 'hello'
  print(lower_string)
  ```

- **`.upper()`**：將字串轉換為大寫。

  ```python
  upper_string = "Hello".upper()  # 'HELLO'
  print(upper_string)
  ```

- **`.strip()`**：移除首尾的空白字符。

  ```python
  whitespace_string = "   Hello   "
  stripped_string = whitespace_string.strip()  # 'Hello'
  print(stripped_string)
  ```

- **`.replace(old, new)`**：將字串中的某個子字串替換為另一個子字串。

  ```python
  replaced_string = "Hello, World!".replace("World", "Python")  # 'Hello, Python!'
  print(replaced_string)
  ```

- **`.split(delimiter)`**：根據分隔符將字串分割成子字串的列表。

  ```python
  split_string = "Hello, World!".split(", ")  # ['Hello', 'World!']
  print(split_string)
  ```

- **`.join(iterable)`**：將可迭代物件（如列表）的元素合併為單一字串，並使用字串作為分隔符。

  ```python
  words = ['Hello', 'World']
  joined_string = " ".join(words)  # 'Hello World'
  print(joined_string)
  ``` 

````python
# 字串創建
greeting = "Hello, World!"

# 字串連接
name = "Alice"
full_greeting = greeting + " " + name
print(full_greeting)  # Hello, World! Alice

# 字串方法
print(greeting.lower())  # hello, world!
print(greeting.upper())  # HELLO, WORLD!
print(greeting.replace("World", "Python"))  # Hello, Python!

# 字串長度
length = len(greeting)
print("問候語的長度:", length)  # 問候語的長度: 13 
````
